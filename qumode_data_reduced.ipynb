## Prerequisite

Packages needed for plotting.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Data

### Binary knapsack problem

```python
bkp_values = [2, 5, 7, 3]
bkp_weights = [2.5, 3, 4, 3.5]
bkp_max_weight = 7

bkp_l_val = 2
bkp_gs_en = -12.0
```

```python
# Read data from file:
datafile = np.load('Qumode-BKP-Results.npz')

# Unpack the datafile to get the necessary variables
bkp_N1 = datafile['bkp_N1']
bkp_nd5_iters = datafile['bkp_nd5_iters']
bkp_results_nd5_en = datafile['bkp_results_nd5_en']
bkp_results_nd5_param = datafile['bkp_results_nd5_param']
bkp_probs_nd5_niter10 = datafile['bkp_probs_nd5_niter10']
bkp_probs_nd5_niter20 = datafile['bkp_probs_nd5_niter20']
bkp_probs_nd5_niter40 = datafile['bkp_probs_nd5_niter40']
bkp_probs_nd5_niter60 = datafile['bkp_probs_nd5_niter60']
bkp_probs_nd5_niter70 = datafile['bkp_probs_nd5_niter70']
bkp_probs_nd5_niter80 = datafile['bkp_probs_nd5_niter80']
bkp_results_nd5 = list(zip(bkp_results_nd5_en, bkp_results_nd5_param))
```

### Multiple constraints

```python
# Read data from file:
multiconstraint_data = np.load('Qumode-QUBO-Multiple-Constraints.npz')

# Unpack the datafile to get the necessary variables
mc_lambda_vals = multiconstraint_data['mc_lambda_vals']
mc_N1 = multiconstraint_data['mc_N1']
mc_nd10_iters = multiconstraint_data['mc_nd10_iters']
mc_results_nd10_ens = multiconstraint_data['mc_results_nd10_ens']
mc_results_nd10_params = multiconstraint_data['mc_results_nd10_params']
mc_probs_nd10_niter10 = multiconstraint_data['mc_probs_nd10_niter10']
mc_probs_nd10_niter20 = multiconstraint_data['mc_probs_nd10_niter20']
mc_probs_nd10_niter30 = multiconstraint_data['mc_probs_nd10_niter30']
mc_probs_nd10_niter40 = multiconstraint_data['mc_probs_nd10_niter40']
mc_probs_nd10_niter50 = multiconstraint_data['mc_probs_nd10_niter50']
mc_probs_nd10_niter60 = multiconstraint_data['mc_probs_nd10_niter60']
mc_probs_nd10_niter70 = multiconstraint_data['mc_probs_nd10_niter70']
mc_probs_nd10_niter80 = multiconstraint_data['mc_probs_nd10_niter80']
mc_probs_nd10_niter90 = multiconstraint_data['mc_probs_nd10_niter90']
mc_probs_nd10_niter100 = multiconstraint_data['mc_probs_nd10_niter100']
mc_probs_nd10_niter200 = multiconstraint_data['mc_probs_nd10_niter200']

mc_results_nd10 = list(zip(mc_results_nd10_ens, mc_results_nd10_params))
```

### BKP with noise

```python
# Load data from file:
bkp_noise_data = np.load('Qumode-BKP-Noise-Results.npz')
# Unpack data to appropriate variable names:
bkp_noise_probs_nd5_k1em1_niter80 = bkp_noise_data['bkp_noise_probs_nd5_k1em1_niter80']
bkp_noise_probs_nd5_k1em2_niter80 = bkp_noise_data['bkp_noise_probs_nd5_k1em2_niter80']
bkp_noise_probs_nd5_k1em3_niter80 = bkp_noise_data['bkp_noise_probs_nd5_k1em3_niter80']
bkp_noise_results_nd5_k1em1_ens = bkp_noise_data['bkp_noise_results_nd5_k1em1_ens']
bkp_noise_results_nd5_k1em1_params = bkp_noise_data['bkp_noise_results_nd5_k1em1_params']
bkp_noise_results_nd5_k1em2_ens = bkp_noise_data['bkp_noise_results_nd5_k1em2_ens']
bkp_noise_results_nd5_k1em2_params = bkp_noise_data['bkp_noise_results_nd5_k1em2_params']
bkp_noise_results_nd5_k1em3_ens = bkp_noise_data['bkp_noise_results_nd5_k1em3_ens']
bkp_noise_results_nd5_k1em3_params = bkp_noise_data['bkp_noise_results_nd5_k1em3_params']
bkp_noise_results_nd5_k1em4_ens = bkp_noise_data['bkp_noise_results_nd5_k1em4_ens']
bkp_noise_results_nd5_k1em4_params = bkp_noise_data['bkp_noise_results_nd5_k1em4_params']

bkp_noise_results_nd5_k1em1 = list(zip(bkp_noise_results_nd5_k1em1_ens, bkp_noise_results_nd5_k1em1_params))
bkp_noise_results_nd5_k1em2 = list(zip(bkp_noise_results_nd5_k1em2_ens, bkp_noise_results_nd5_k1em2_params))
bkp_noise_results_nd5_k1em3 = list(zip(bkp_noise_results_nd5_k1em3_ens, bkp_noise_results_nd5_k1em3_params))
bkp_noise_results_nd5_k1em4 = list(zip(bkp_noise_results_nd5_k1em4_ens, bkp_noise_results_nd5_k1em4_params))
```

## Plots

### BKP

```python
# Extracting x and y values
x_values = bkp_nd5_iters.copy()
y_values = np.array([item[0] for item in bkp_results_nd5])

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, marker='o', linestyle='-', color='green')

# Adding a horizontal line at y = -11.5
plt.axhline(y=-12.0, color='black', linestyle='--', label='y = -11.5')

# Adding labels and title
plt.xlabel('Iteration', fontsize=16)
plt.ylabel('Trial energy (a.u.)', fontsize=16)

# Save
#plt.savefig('bkp_energies_lval2_nd5.pdf', dpi=500)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Labels
nvec_labels = [fr'$| {x[0]}, {x[1]}, {x[2]} \rangle $' for x in bkp_N1]
x = np.arange(len(nvec_labels))

# Determine midpoint
midpoint = len(x) // 2

fig = plt.figure(figsize=(28, 16))

# Top subplot: Left half
ax_top = fig.add_subplot(2,1,1)
ax_top.bar(x[:midpoint], bkp_probs_nd5_niter10[:midpoint], color='brown', alpha=0.9, label=r'Iterations = 10', width=1.4)
ax_top.bar(x[:midpoint], bkp_probs_nd5_niter20[:midpoint], color='blue', alpha=0.9, label=r'Iterations = 20', width=1.1)
ax_top.bar(x[:midpoint], bkp_probs_nd5_niter40[:midpoint], color='grey', alpha=0.9, label=r'Iterations = 40', width=0.8)
ax_top.bar(x[:midpoint], bkp_probs_nd5_niter60[:midpoint], color='orange', alpha=0.9, label=r'Iterations = 60', width=0.5)
ax_top.bar(x[:midpoint], bkp_probs_nd5_niter80[:midpoint], color='green', alpha=0.9, label=r'Iterations = 80', width=0.2)
ax_top.set_xticks(x[:midpoint])
ax_top.set_xticklabels(nvec_labels[:midpoint], rotation=90, ha='center', fontsize=16)
ax_top.tick_params(axis='x', pad=10)
ax_top.tick_params(axis='y', labelsize=20)
ax_top.set_xlabel('Basis states', fontsize=30)
ax_top.set_ylabel(r'$ | \langle q, n, m | \psi \rangle |^2 $', fontsize=30)
ax_top.axhline(y=1.0, color='grey', linestyle='--', linewidth=2)
ax_top.legend(fontsize=20)

# Bottom subplot: Right half
ax_bottom = fig.add_subplot(2,1,2)
ax_bottom.bar(x[midpoint:], bkp_probs_nd5_niter10[midpoint:], color='brown', alpha=0.9, label=r'Iterations = 10', width=1.4)
ax_bottom.bar(x[midpoint:], bkp_probs_nd5_niter20[midpoint:], color='blue', alpha=0.9, label=r'Iterations = 20', width=1.1)
ax_bottom.bar(x[midpoint:], bkp_probs_nd5_niter40[midpoint:], color='grey', alpha=0.9, label=r'Iterations = 40', width=0.8)
ax_bottom.bar(x[midpoint:], bkp_probs_nd5_niter60[midpoint:], color='orange', alpha=0.9, label=r'Iterations = 60', width=0.5)
ax_bottom.bar(x[midpoint:], bkp_probs_nd5_niter80[midpoint:], color='green', alpha=0.9, label=r'Iterations = 80', width=0.2)
ax_bottom.set_xticks(x[midpoint:])
ax_bottom.set_xticklabels(nvec_labels[midpoint:], rotation=90, ha='center', fontsize=16)
ax_bottom.tick_params(axis='x', pad=10)
ax_bottom.tick_params(axis='y', labelsize=20)
ax_bottom.set_xlabel('Basis states', fontsize=30)
ax_bottom.set_ylabel(r'$ | \langle q, n, m | \psi \rangle |^2 $', fontsize=30)
ax_bottom.axhline(y=1.0, color='grey', linestyle='--', linewidth=2)
ax_bottom.legend(fontsize=20)

# For the top subplot (left half):
ax_top.set_xlim(-0.4, midpoint - 0.4)  # This ensures the first bar starts near the left edge

# For the bottom subplot (right half):
ax_bottom.set_xlim(midpoint - 0.4, len(x) - 0.4)

plt.tight_layout(pad=4.0)
plt.subplots_adjust(bottom=0.2)

# Save the plot with higher DPI to ensure quality
#plt.savefig('bkp_probs_nd5_niters.pdf', dpi=500)

plt.show()
```
